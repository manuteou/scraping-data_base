```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
urls= ['https://www.snrbaudouin.com/avant-l-art-moderne-c102x3977606',
       'https://www.snrbaudouin.com/art-moderne-art-contemporain-c102x3977610',
       'https://www.snrbaudouin.com/artistes-modernes-et-contemporains-c102x3977605',
       'https://www.snrbaudouin.com/architecture-c102x3977621',
       'https://www.snrbaudouin.com/arts-decoratifs-arts-appliques-c102x3977604',
       'https://www.snrbaudouin.com/photographie-c102x3977602',
       'https://www.snrbaudouin.com/galerie-pierre-matisse-c102x3977712',
       'https://www.snrbaudouin.com/reperes-galerie-lelong-c102x3977675',
       'https://www.snrbaudouin.com/arts-du-spectacle-et-musique-c102x3977609',
       'https://www.snrbaudouin.com/mode-textiles-c102x3977607',
       'https://www.snrbaudouin.com/cheval-d-attaque-c102x3977648',
       'https://www.snrbaudouin.com/le-dernier-terrain-vague-c102x3977665',
       'https://www.snrbaudouin.com/paris-france-c102x3977612',
       'https://www.snrbaudouin.com/esoterisme-c102x3977611',
       'https://www.snrbaudouin.com/jeunesse-et-bandes-dessinees-c102x3977616']
```

```python
df_ebay = pd.read_csv('FileExchange.csv', sep=';', encoding='latin')
```

```python
df = df_ebay["Titre de l'objet"].str.split(" - ", expand=True, n=1)
df.columns = ['titre','nom']
df['titre'] = df['titre'].apply(lambda x: x.replace("."," ").replace(","," ").replace("!"," ").replace(")"," ").replace("("," ").replace("[\]"," ").replace('"',' ').replace("/"," "))
df['titre'] = df['titre'].str.replace(r" +"," ")
df['titre'] = df['titre'].str.lower()
```

```python
results = []
for url in urls:
    print("collecting url:",url)
    tempo = []
    reponse = requests.get(url)
    soup = BeautifulSoup(reponse.text, 'html.parser')
    titre = soup.findAll("h3")
    for t in titre:
        tempo.append(t.get_text())
    for i in range(len(tempo)):
        tempo[i] = tempo[i].replace("""<h3 class="PBMainTxt">""","")
        tempo[i] = tempo[i].replace("""</h3>""","")
        tempo[i] = tempo[i].lower()
        tempo[i] = tempo[i].split("-")[0]
        tempo[i] = tempo[i].replace("."," ").replace(","," ").replace("!"," ").replace(")"," ").replace("("," ").replace("[\]"," ").replace('"',' ').replace("/"," ")
        tempo[i] = re.sub(" +"," ",tempo[i])
        tempo[i] = tempo[i].strip()
        try :
            tempo[i] = tempo[i].split(']')[1]
        except(IndexError):
            print('jump')
    results.append(tempo)
```

```python
remove = []
for l in range(len(results)):
    remove.append(list(df['titre'][df['titre'].isin(results[l])].values))
    
```

```python
final = []
for i,elt in enumerate(remove):
    for li in elt:
        final.append(li)
```

```python
result_final = []
for i,elt in enumerate(results):
    for li in elt:
        result_final.append(li)
```

```python
enfin = []
for elt in result_final:
    for l in final:
        if elt == l:
            enfin.append(elt)
print('Nombres de livres à rajouter à la base:', len(enfin))
```
